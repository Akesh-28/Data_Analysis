```python
import pandas as pd
import numpy as nnp
```

READING CSV

```python
df=pd.read_csv("sales_data_sample.csv", encoding='latin1')
total=pd.read_csv("sales_data_sample.csv", encoding='latin1')
```

#Updating columns

```python
df.columns=["ID","QUANTITY","PRICE/Qty","LINE NUMBER","SALES","DATE","STATUS","QUARTERS","MONTH","YEAR","PRODUCT GROUP","MSRP","CODE","CUSTOMER","PHONE","ADDRESS","ADDRESSLINE2","CITY","STATE","POSTAL CODE","COUNTRY","TERRITORY","FIRST CONTACT","LAST CONTACT","DEAL SIZE"]
df
```

UPDATED ADDRESS

```python
if 'ADDRESSLINE2' in df.columns:
    df['ADDRESS'] = df['ADDRESS'].astype(str) + ', ' + df['ADDRESSLINE2'].fillna('').astype(str)
    df.drop(columns=["ADDRESSLINE2"], inplace=True)
df
```

UPDATING DATE(from object) AND REMOVE TIME FROM DATE(since all dates have same time)

```python
df["DATE"]=pd.to_datetime(df["DATE"],format="%m/%d/%Y %H:%M")
df["DATE"]=pd.to_datetime(df["DATE"]).dt.date
```

SORTED BY

```python
df.sort_values(by=["DATE"],ascending=[True],inplace=True)
```

```python
df.nunique()
```

UPDATING CORRECT CALCULATIONS IN wrong rows

```python
wrong_rows = df[df["SALES"] != df["PRICE/Qty"] * df["QUANTITY"]]
wrong_rows
```

```python
df["SALES"]=df["PRICE/Qty"]*df[("QUANTITY")]
df
```

TOP BUYERS

```python
hwcust=df.groupby("CUSTOMER").agg(TOTAL_SALES=('SALES','sum'),UNIQUE_IDS=("ID", "nunique") ).reset_index()
hwcust.sort_values("TOTAL_SALES",ascending=False)
```

GROUPING BASED ON STATUS

```python
status_group=dict(tuple(df.groupby("STATUS")))
```

```python
shipped=status_group["Shipped"]
resolved=status_group["Resolved"]
cancelled=df[df["STATUS"]=="Cancelled"]
on_hold=df.query("STATUS=='On Hold'")
disputed=df[df["STATUS"]=="Disputed"]
in_process=df[df["STATUS"]=="In Process"]
```
