**LIBRARIES USED**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

**READING THE SOURCE FILE**

```python
df=pd.read_csv("Final_Project_Pandas2.csv",encoding='latin')
df
```

**CONTINOUS:**

**SALES ANALYTICS**

Quantity vs Country

```python
plt.figure(figsize=(20,10))
plt.bar(df["COUNTRY"],df["QUANTITY"],width=0.4)
plt.show()
```

Sales by year:

```python
figure, axes = plt.subplots(1, 2, figsize=(6,6))
years = df["YEAR"].value_counts().head().sort_index()
Total = df.groupby("YEAR")["SALES"].sum().sort_index()
colors = ["skyblue", "orange", "green"]

axes[0].pie(years, labels=years.index, autopct="%1.1f",colors=colors)
axes[1].pie(Total, labels=Total.index, autopct="%1.1f",colors=colors)
axes[0].set_title("TOP BY QUANTITY ")
axes[1].set_title("TOP BY SALES")
plt.show()
```

**DISCRETE:**

*Random Data*

```python
scores=np.random.normal(loc=80,scale=10,size=100)
scores=np.clip(scores,0,100)
plt.hist(scores,bins=10,color="green")
plt.title("SCORES")
plt.xlabel("scores")
plt.ylabel("no of students")
```

```python
plt.figure(figsize=(12,6))
df.boxplot(column="QUANTITY", by="COUNTRY", rot=90)
plt.title("Quantity by Country")
plt.suptitle("")
plt.show()
```

```python
df_unique = df.drop_duplicates(subset=["CUSTOMER", "TOTAL_SALES"])
tot=df_unique.groupby(["DEAL SIZE","STATUS"])["TOTAL_SALES"].sum().reset_index()
tot["TOTAL_SALES"] = tot["TOTAL_SALES"] / 1000000
sns.barplot(data=tot,x="DEAL SIZE",y="TOTAL_SALES",hue="STATUS")
sns.set_context("paper")
plt.ylabel("Total Sales (Millions)")
```

```python
sns.set_style("whitegrid")
sns.pairplot(df[["SALES","QUANTITY","PRICE/Qty","MSRP",
                "DEAL SIZE"]],hue="DEAL SIZE")
```

```python
sns.countplot(data=df,x="PRODUCT GROUP",hue="DEAL SIZE")
plt.xticks(rotation=45, fontsize=10)
```

```python
qt=df.groupby(["QUARTERS","YEAR"])["SALES"].sum().reset_index()
qt["SALES"] = qt["SALES"] / 1000000
sns.barplot(data=qt,x="QUARTERS",y="SALES",hue="YEAR")
plt.ylabel("Total Sales (Millions)")
```

```python
top3 = df.groupby("CUSTOMER")["SALES"].sum().nlargest(3).index
top = df[df["CUSTOMER"].isin(top3)]
sns.swarmplot(data=top,x="CUSTOMER",y="PRICE/Qty",hue="PRODUCT GROUP")
plt.xticks(rotation=45,fontsize=8)
plt.ylim(30,200)
```
